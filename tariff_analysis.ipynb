### Шаг 2. Подготовка данных
Изменим тип данных на представление даты:

```python
users['reg_date'] = pd.to_datetime(users['reg_date'], format='%Y-%m-%d')
calls['call_date'] = pd.to_datetime(calls['call_date'], format='%Y-%m-%d')
messages['message_date'] = pd.to_datetime(messages['message_date'], format='%Y-%m-%d')
internet['session_date'] = pd.to_datetime(internet['session_date'], format='%Y-%m-%d')
```

Удалим столбец internet[Unnamed: 0]

```python
internet = internet.drop('Unnamed: 0', axis=1)
```

Поменяем тип данных internet[mb_used] и calls[duration] на int и округлим в большую сторону 

```python
internet['mb_used'] = np.ceil(internet['mb_used']).astype('int')
calls['duration'] = np.ceil(calls['duration']).astype('int')
```

Добавим столбец месяца подключения тарифа для пользователя:

```python
def add_month(data, column):
    data['month'] = pd.DatetimeIndex(data[column]).month
add_month(users, 'reg_date')
add_month(calls, 'call_date')
add_month(messages, 'message_date')
add_month(internet, 'session_date')
```

Удалим пустые значения из столбца calls['duration']

```python
calls = calls.query('duration != 0')
```

Посчитаем количество сделанных звонков и израсходованных минут разговора по месяцам для каждого пользователя:
- количество звонков

```python
calls_count = calls.pivot_table(index=['user_id', 'month'], values='id', aggfunc='count')
calls_count.head()
```

- количество израсходованных минут разговора:

```python
calls_duration = calls.pivot_table(index=['user_id', 'month'], values='duration', aggfunc='sum')
calls_duration.head()
```

Объединим эти сводные таблицы в одну методом merge, а также изменим названия столбцов количества звонков и количества минут:

```python
calls_group = calls_count.merge(calls_duration, on=['user_id', 'month'], how='left').reset_index()
calls_group = calls_group.rename(columns={'id': 'number_of_calls'})
calls_group.head()
```

Посчитаем количество отправленных сообщений по месяцам:

```python
messages_count = messages.pivot_table(index=['user_id', 'month'], values='id', aggfunc='count').reset_index()
messages_count = messages_count.rename(columns={'id': 'number_of_messages'})
messages_count.head()
```

Посчитаем объем израсходованного интернет-трафика по месяцам:

```python
mb_used_sum = internet.pivot_table(index=['user_id', 'month'], values='mb_used', aggfunc='sum').reset_index()
mb_used_sum = mb_used_sum.rename(columns={'mb_used': 'mb_used_month'})
mb_used_sum.head()
```

Объединим все таблицы в одну, заменим названия некоторых столбцов с месяцами:

```python
df = calls_group.merge(messages_count, on=['user_id', 'month'], how='outer')
```

```python
df = df.merge(mb_used_sum, on=['user_id', 'month'], how='outer')

df = df.merge(users, on='user_id', how='left')
df = df.rename(columns={'month_y': 'tariff_connect_month', 'month_x': 'month'})
df.head()
```

```python
df.info()
```

При объединении таблиц возникли пропуски данных. Они могли возникнуть по причине того, что пользователь в какой-либо месяц не пользовался интернетом или сообщениями. Заменим пропуски нулями, чтобы при расчете выручки не было пропусков.

```python
columns = ['number_of_calls', 'duration', 'number_of_messages', 'mb_used_month']
for column in columns:
    df[column] = df[column].fillna(0)
```

```python
df.info()
```

Напишем функцию для расчета выручки:

```python
tariffs
```

```python
def proceeds(row):
    proceeds = 0
    if row['tariff'] == 'smart':
        duration_balance = row['duration'] - tariffs.loc[0]['minutes_included']
        if duration_balance > 0:
            proceeds += duration_balance * 3
        messages_balance = row['number_of_messages'] - tariffs.loc[0]['messages_included']
        if messages_balance > 0:
            proceeds += messages_balance * 3
        mb_balance = row['mb_used_month'] - tariffs.loc[0]['mb_per_month_included']
        if mb_balance > 0:
            proceeds += (mb_balance / 1024) * 200
        if proceeds > 0:
            proceeds += 550
        else:
            proceeds = 550
    else:
        duration_balance = row['duration'] - tariffs.loc[1]['minutes_included']
        if duration_balance > 0:
            proceeds += duration_balance * 1
        messages_balance = row['number_of_messages'] - tariffs.loc[1]['messages_included']
        if messages_balance > 0:
            proceeds += messages_balance * 1
        mb_balance = row['mb_used_month'] - tariffs.loc[1]['mb_per_month_included']
        if mb_balance > 0:
            proceeds += (mb_balance / 1024) * 150
        if proceeds > 0:
            proceeds += 1950
        else:
            proceeds = 1950
    return proceeds
```

```python
df['proceeds'] = df.apply(proceeds, axis=1)
```

```python
df.head()
```

Узнаем, сколько раз пользователи превышали тарифные минуты, сообщения и трафик:

```python
print('Количество превышений на тарифе smart:')
print(len(df.query('tariff == "smart" and proceeds > 550')))
print()
print('Количество превышений на тарифе ultra:')
print(len(df.query('tariff == "ultra" and proceeds > 1950')))
```

### Шаг 4. Проверка гипотез
1. средняя выручка пользователей тарифов «Ультра» и «Смарт» различается

- H0 - средняя выручка тарифов «Ультра» и «Смарт» не различается
- H1 - средняя выручка тарифов «Ультра» и «Смарт» различается

Если критерий значимости выше 5%, нулевая гипотеза не отвергается.

Методом теста Левена оценим равенство дисперсий, используя p-уровень значимости для проверки соблюдение допущения о гомогенности дисперсий. Тест Левена предполагает, что выборки ненормально распределены. Проверим нормальность с помощью теста Шапиро-Уилка. Если по результатам применения теста мы видим, что нормальность соблюдается, мы используем для проверки равенства (гомогенности) дисперсий тест Бартлета (проверка на равенство дисперсий двух нормально распределенных выборок). В противном случае - тест Левена.

Проведем статистический тест и применим метод scipy.stats.ttest_ind (array1, array2, equal_var). 

```python
## Проверяем, как распределены выборки
shapiro_test_ultra = st.shapiro(df_ultra['proceeds'])
shapiro_test_smart = st.shapiro(df_smart['proceeds'])
print('Результаты теста Шапиро, тариф ultra:', shapiro_test_ultra)
print('Результаты теста Шапиро, тариф smart:', shapiro_test_smart)
```

pvalue в тарифе smart и ultra меньше, чем alpha=0.05. Отклоняем нулевую гипотезу. Нормальность не соблюдается, поэтому с помощью теста Левена оценим равенство дисперсий, используя p-уровень значимости.

```python
st.levene(df_ultra['proceeds'], df_smart['proceeds'], center='mean')
```

p-value меньше заданного 0.05, значит выборки гетерогенны. В st.ttest_ind параметр
equal_var=False.

```python
alpha = 0.05
results = st.ttest_ind(df_smart['proceeds'], df_ultra['proceeds'], equal_var=False)
print('p-значение:', results.pvalue)
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')  
```

Средняя выручка пользователей тарифов «Ультра» и «Смарт» различается.

- Н1 - средняя выручка пользователей из Москвы и других регионов различается
